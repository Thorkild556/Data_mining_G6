```python
from src.cluster_metrics import TrialAndError
from concurrent.futures.thread import ThreadPoolExecutor
from datasets import load_dataset
import plotly.express as px
```

```python
# Load dataset
dataset = load_dataset("mteb/twentynewsgroups-clustering", revision="6125ec4e24fa026cec8a478383ee943acfbd5449")
print(dataset["test"])

# Extracting variables
test = dataset["test"]
texts_sep = list(test["sentences"])
labels_sep = list(test["labels"])
```

```python
trailer = TrialAndError(dataset=dataset)
trailer.eval_db_scan()
```

```python
result = []

threads = []
with ThreadPoolExecutor(max_workers=6) as executor:
    threads = executor.map(lambda grp_size: TrialAndError(dataset=dataset, grp=grp_size), range(len(texts_sep)))

with ThreadPoolExecutor(max_workers=6) as executor:
    result = list(executor.map(lambda thread: thread.eval_db_scan(), threads))

print(result)
```
